```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pidmat = pd.read_csv('unique_pidmat.csv', index_col=0)
pidmat = pidmat.to_numpy()
plt.imshow(pidmat)
plt.show()
pidmat
```

```python
### histograms ###
max_pid = np.sort(pidmat, axis=1)[:,::-1]
plt.plot(max_pid[1,:], color='k')
plt.plot(max_pid[20,:], color='r')
plt.plot(max_pid[200,:], color='g')
plt.plot(max_pid[1200,:], color='b')
plt.show()

mean_pid = np.mean(pidmat, axis=0)
plt.hist(mean_pid, color='k', bins=50)
plt.show()

plt.hist(pidmat.flatten(), color='k', bins=50)
plt.show()

max_pid
max_pid[0,:]
```
